```python
import impectPy as ip
import pandas as pd
import numpy as np
import json
import mplsoccer
from pathlib import Path
```

```python
creds_path = Path("creds") / "impect.json"
with creds_path.open() as f:
    impect_creds = json.load(f)

username = impect_creds["username"]
password = impect_creds["pswd"]

# get access token
token = ip.getAccessToken(username=username, password=password)
```

```python
# get list of iterations
iterations = ip.getIterations(token=token)

# print iterations to console
iterations
```

```python
# get matches for iteration
matchplan = ip.getMatches(iteration=1217, token=token)

# print matches to console
matchplan
```
